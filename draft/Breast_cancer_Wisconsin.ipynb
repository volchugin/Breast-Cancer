```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

sns.set()
```

```python
df = pd.read_csv ('/Users/denisvolchugin/Github/Breast-cancer/data/data.csv')
df.head()
```

```python
df.info()
```

```python
df.drop (['Unnamed: 32', 'id'], inplace=True, axis=1)
```

```python
df.describe()
```

```python
y = df.diagnosis
```

```python
data = df.iloc[:, 1:]
```

```python
StData = StandardScaler().fit_transform(data)
```

```python
pca = PCA(.90)
```

```python
PCs = pca.fit_transform(X = StData)
print (pca.n_components_)
```

```python
principalDf = pd.DataFrame(data = PCs,
                          columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7'])
```

```python
principalDf.head()
```

```python
fDf = pd.concat([principalDf, y], axis = 1)
```

```python
g = sns.pairplot(fDf, hue="diagnosis", palette="Set2", diag_kind="kde", height=2.5)
```
